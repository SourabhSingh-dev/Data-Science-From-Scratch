```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], kde = True, bins = 20)
plt.subplot(1,2,2)
sns.histplot(df['tip'], kde = True, bins = 20)
```

```python
sns.jointplot(x = df['total_bill'], y = df['tip'],data = df, kind = 'scatter')
```

```python
sns.jointplot(x = df['total_bill'], y = df['tip'],data = df, kind = 'kde')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```
