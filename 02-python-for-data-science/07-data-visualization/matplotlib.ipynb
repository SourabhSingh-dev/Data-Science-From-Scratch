```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.arange(0,20,2)
arr2 = arr1 ** 2
plt.plot(arr1,arr2)
plt.title("Hello")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(arr1,arr2)
plt.subplot(2,2,2)
plt.plot(arr1,arr1)
plt.subplot(2,2,3)
plt.plot(arr1,arr2**2)
plt.subplot(2,2,4)
plt.plot(arr1,arr1**3)
```

```python
fig = plt.figure(figsize=(20,8))
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(arr1,arr2)
axis2 = fig.add_axes([0.5,0.4,0.2,0.4])
axis2.plot(arr1,arr1)
```
