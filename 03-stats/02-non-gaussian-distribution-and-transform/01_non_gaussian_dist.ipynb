```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../01-pdf-in-ds/iris.csv')
df.head()
```

```python
sns.kdeplot(data = df, x = 'sepal_length')
```

```python
temp = sorted(df['sepal_length'])
```

```python
samples = np.random.normal(loc = 0, scale = 1, size = 1000)
```

```python
theo_quant = []
normal_quant = []
for i in range(1,101):
    theo_quant.append(np.percentile(samples,i))
    normal_quant.append(np.percentile(temp,i))
```

```python
sns.scatterplot(x = theo_quant, y = normal_quant)
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
fig = sm.qqplot(df['sepal_length'], line = '45' , fit = True )
```

```python
alpha = 3
xm = 1
x = np.linspace(1, 10, 1000)
y = (alpha * (xm ** alpha))/(x**(alpha + 1))
plt.plot(x,y)
```

```python
plt.plot(np.log(x),np.log(y))
```

```python
import scipy.stats as stats
```

```python
x = stats.pareto.rvs(b=alpha, scale= xm, size = 1000)
#floc = 0 , do not move the distribution left or right
params = stats.pareto.fit(x, floc= 0)

dist = stats.pareto(b=params[0],scale=params[2])
```

```python
fig = sm.qqplot(x, dist, line= '45')
```
